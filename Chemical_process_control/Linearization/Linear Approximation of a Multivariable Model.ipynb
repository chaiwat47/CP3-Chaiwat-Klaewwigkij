```python
#Simulation of a Gravity-Drained Tank
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
#Parameter values
qin = 1
A = 1
C = 2
IC =[0.0]
#steady state
hbar = (qin/C)**2

#nonlinear simulation 
def hdot(h,t):
    dhdt = (qin-C*np.sqrt(h))/A
    return dhdt
## x,y axis
t = np.linspace(0,5)
h = odeint(hdot, IC,t)

#linear approximation
a = -C**2/(2*A*qin)
b = 1/A

def xdot(x,t):
    dxdt = a*x+b*u
    return dxdt
u = 0
x = odeint(xdot, [-hbar], t) #IC = [-hbar]

#visualization
plt.plot(t,h)
plt.plot(t,hbar+x)
legend_graph = ['Nonlinear Simulation','Linear Approximation']
plt.legend(legend_graph,loc = 'lower right')
plt.xlabel('Time [min]')
plt.ylabel('Height [m]')
plt.title('Nonlinear Simulation vs. Nonlinear Simulation') 
plt.grid()
plt.show()
```
