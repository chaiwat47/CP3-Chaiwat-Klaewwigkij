```python
#กราฟ Histrogram ใช้สำหรับนับหรือแสดงความถี่ของข้อมูล เช่น ต้องการพล็อตดูว่าลูกค้าอายุช่วงต่างๆ มีกี่คน
#ข้อมูลชนิดไม่ต่อเนื่อง จะเป็นข้อมูลแบบ Norminal & Ordinalเช่นเก็บข้อมูลลูกค้าที่มาใช้บริการแต่ละช่วงอายุ

age = [18,17,20,18,18,17,19,21,18,17,18,17,18,18,19,19,18,17]

plt.hist(age)
plt.show()

#หากต้องการนับความถี่เพื่อแสดงตัวเลขในแต่ละช่วงอายุ
import pandas as pd
df = pd.DataFrame(age, columns = ['Age']) #สร้าง DataFrame
df.head()
df.Age.value_counts() #นับความพี่

#ข้อมูลชนิดต่อเนื่อง
#สร้างข้อมูลด้วยค่าตัวแปรสุ่ม ที่มีค่าเฉลลี่ยอยู่ที่ 10 ค่าเบี่ยงเบน 2.5 จำนวน 5000 ข้อมูล 
wt =np.random.normal(10,2.5,5000)

plt.hist(wt) #ไม่กำหนดช่วง bins ค่า defualt = 10 (แบ่งเป็น 10 กลุ่ม)
plt.show()
plt.hist(wt, bins = 500) #ยิ่งใช้binsเยอะเวลาในการสร้างกราฟจะนานขึ้น
plt.show()
#การสร้าง Histrogram แบบต่อเนื่องจะเห็นว่าใช้ข้อมูลบในปริมาณที่มาก ดังนั้นในทางปฎิบัติ จึงใช้วิธีการ bins #ลองตรวจดูค่าในข้อมูล wt ซึ่งมีชนิดข้อมูลเป็นarray มี 5000 ข้อมูล แต่เมื่อเราดู 10 ข้อมูลแรก จะพบว่าช่วงขข้อมูลที่8 -13 มีค่าเฉลี่ยอยู่ที่ 10 ตามที่เราสร้าง
wt[:10]
```

# Workshop 10 : กราฟแบบกำหนดรายละเอียด

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6] #ชุดข้อมูล
ydata1 = [1,2.8,2.1,4,3,3.5]
ydata2 = [2.2,1.6,2.3,2,2.3,2.1]

plt.rcParams.update({'font.size':15, 'font.family':'tahoma'}) #ขนาดและรูปแบบอักษร
plt.figure(figsize=(6,4)) #ขนาดกราฟกว้าง * สูง

plt.grid(linestyle = '--', axis = 'y') #แสดงเส้นกริดเฉพาะแกน y
plt.bar(x, ydata1, color = 'c', width =0.5 , label = 'ในประเทศ') # c = cyan
plt.bar([x +0.2 for x in x], #ระยะขยับกราฟแต่ละแท่ง
            ydata2, width = 0.5, color = 'orange' ,alpha=0.9 ,label = 'ส่งออก')

plt.plot(x , ydata1, color = 'b', marker = 's', markersize = 8) # b=blue
plt.plot(x , ydata2, color = 'r', marker = 'o', markersize = 8) # r=red

plt.xlabel("เดือน")
plt.ylabel("ยอดขาย")
plt.title("รายงาน Sales Report")

for tx,ty in list(zip(x, ydata1)):
    plt.annotate(ty,                           # ข้อความกำกับ
                (tx,ty),                       # ตำแหน่ง
                textcoords="offset points",    # ระบุการเยื้องตำแหน่ง offset
                xytext=(0,8),                  # ระยะเยื้องตำแหน่งจาก points (x,y)
                ha = 'center',                 # จัดตำแห่งตรงกลาง
                fontsize = 12)                 # ขนาดอักษร

plt.legend()                                   # แสดงข้อความกำกับ(Legend)
plt.ylim(0,4.5)                                # ช่วงพิกัด y 0-4.5
plt.show()
```

```python
#โหลดข้อมูล
#ตัวอย่างนี้จะโหดข้อมูลประกอบด้วยค่าอาหารและค่าทิปที่ลูกค้าจ่ายให้กับร้าน ทั้งหมดมีจำนวน 244 ใบเสร็จ(โต๊ะ) โดยรายละเอียดคอลัมน์ (ตัวแปร) ในไฟล์มีดังนี้
# total_bill ยอดค่าอาหารแต่ละโต๊ะ
# tip ค่าทิปที่ลูกค้าให้ทางร้าน
# sex เพศของลูกค้า (Male/Female)
# smoker ลูกค้าสูบหรือไม่สูบบุหรี่(Yes / No) อาตเทียบได้กับการบันทึกบริการที่มีอยู่จริงของร้านอาหาร เช่น บริเวณโซนริมน้ำหรือโซนเปิดโล่งสำหรับสูบบุหรี่ และโซนห้องแอร์ไม่บุหรี่
# day วัน
# time มื้ออาหาร
# size ขนาดโต๊ะ หรือพูดง่ายๆ คือจำนวนสมาชิกของแต่ละโต๊ะที่เข้ามารับประทานอาหารแต่ละครั้ง เช่นโต๊ะลูกค้า 2 คน 3 คน
import seaborn as sns
import pandas as pd

#matplotlib ไม่ต้องโหลดก็ได้ pd มีแล้ว
# แต่ในตัวอย่างนี้ใช้เขียน Title และ save รูป ดังนั้นจึงต้องimport matplotlib
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')     # โหลดข้อมูลจากเว็บ

#ตรวจสอบข้อมูลเบื่องต้น : ดู head tail sample ข้อมูลเบื้องต้น ดูชนิด ดูขนาด
df.head()
df.tail()
df.sample(5) #สุ่มข้อมูล 5 แถว

#ตรวจดูชนิดข้อมูล
df.dtypes #ชนิด
# df.info()
# df.shape ขนาดข้อมูล 
df.describe() # คำนวณหาค่าสถิติเบื้องต้น
df['size'] = df['size'].astype('category') #เปลี่ยน size จากข้อมูล int ป็น category
df.dtypes
#ตรวจสอบข้อมูลที่หาย
df.isnull().sum() #หา missing data ช่องไหนคือ 0 คือไม่มีmissing data
#ความถี่ของขนาดโต๊ะ hist ใน seaborn
df['size'].hist()
plt.show()
#sns.countplot(df['size']) #กราฟของseaborn จะให้สีลงไปด้วย
#sns.countplot('size', data =df)
sns.distplot(df['size'], kde = False)
plt.savefig('chart1.png') #การบันทึกไฟล์

#ในกรณีที่ต้องการตัวเลขสรุปความถี่ ใช้โค้ดดังนี้
#df['size'].value_counts()
# df['size'].value_counts().to_frame(name='freq') # .to_frame(name='freq')จะได้เก็บเป็น DataFrame 
# pd.DataFrame(df['size'].value_counts()) # จะได้เก็บเป็น DataFrame

#ในกรณีที่ใช้ Pivot Table จะได้ DataFrame ที่มีการแจกแจงความถี่เท่ากัน
#pvt = df.pivot_table(index = 'size', values='total_bill', aggfunc = 'count') # of len(df)
#pvt=pvt.sort_values(by='total_bill',ascending = False) #ถ้าเรียงจากมากไปน้อย
#pvt

#สรุปคือร้านควรพิจารณาจัดโต๊ะรองรับสำหรับสองคนให้มากที่สุด รองลงไป คือโต๊ะ 3-4 ที่นั่ง จึงจะรองรับจำนวนคนได้มากสุด

#การตกแต่งกราฟ
# พื้น + เส้นตาราง
#sns.set_style("whitegrid") #พื้นขาว มีเส้นตาราง
sns.set_style("darkgrid") #พื้นเทา มีเส้นตาราง
#sns.set_style("white") #พื้นขาว ไม่มีเส้นตาราง
#sns.set_style("datk") #พื้นเทา ไม่มีเส้นตาาราง

# ขนาดภาพ และ ข้อความกำกับ
plt.figure(figsize=(8,5))
plt.rcParams.update({'font.size':16,'font.family':'tahoma'})
plt.title('กราฟ Histogram - Tip')
#plt.legend() #ต้องระบุ (x,y)
df['size'].hist()
plt.show()

plt.title('ลูกค้า-Day : sorting')
sns.countplot('day',data=df, order = ['Sat','Sun','Thur','Fri']) #orderต้องเป็นชื่อคอลัมน์ที่เก็บไว้ด้วย
plt.show()

#ขนาดโต๊ะรายวัน
#จัดโต๊ะให้เหมาะสมรายวัน
pvt = df.pivot_table(index ='day', columns ='size', values= 'total_bill',aggfunc='count')

pvt.plot.bar()
plt.title('ขนาดโต๊ะ - Day')

plt.legend(loc='upper center')

plt.show()

pvt = df.pivot_table(index='size', columns='day', values = 'total_bill',aggfunc=['count']) #transpostจากด้านบน

pvt = df.pivot_table(index ='day', columns='size', values = 'total_bill',aggfunc=['count']) #transpost

pvt.T #transpost
pvt.columns = pvt.columns.droplevel()
pvt

pvt.plot.bar()
plt.title('ขนาดโต๊ะ - Day')

plt.legend(loc='upper center')

plt.show()

#สรุปจากราจะได้ว่า ต้องการโต๊ะแบบ 2 ที่นั่งมากที่สุด โดยวันเสาร์ควรจัดโต๊ะ 2 ที่นั่ง ให้มั 2-3เท่าของ โต๊ะ 3ที่นั่ง

#วันไหนมีลูกค้ามากที่สุด
# ใช้การนับจำนวนโต๊ะ(จำนวนbill) แบ่งตามวัน ใช้ sns.countplot()
sns.set_style("whitegrid")
plt.figure(figsize = (8,4)) # ถ้าต้องการกำหนดขนาดรูป (ขนาดกราฟ)
plt.rcParams.update({'font.size':16, 'font.family':'tahoma'}) #ขนาดและแบบอักษร
plt.title('ลูกค้า-Day') # หัวข้อกราฟ
sns.countplot('day',data =df) #นับตามวัน
plt.show()

#จะเห็นได้ว่าลูกค้าจะเยอะเสาอาทิตย์ ฝ่ายบริหารจัดการจะต้องเพิ่มพนักงานดูแลเสา-อา มากขึ้นให้หยุด วันธรรมดา
#ตารางแสดงตัวเลขแสดงได้ 2 แบบ
df['day'].value_counts() #แบบที่ 1
df['day'].value_counts().to_frame(name='freq').sort_values('freq',ascending =False)

#ยอดค่าอาหารเฉลี่ย วันไหนมากสุด
plt.title('ค่าอาหาร Bill-Day')
sns.barplot('day','total_bill', data=df) #จะมีเส้น std มาด้วยแสดงถึงการกระจายตัว
plt.show()

#ใช้ lineplot แสดงถึง Confidence interval : CI บ่งบอกถึงการกระจายมากน้อย
plt.title('ค่าอาหาร Bill=Day')
sns.lineplot('day','total_bill',data=df)
#sns.lineplot('day','total_bill',data=df,ci=None) #ci= none ไม่แสดงช่วงแรงเงา\
#ข้อสังเกตแม้ว่าวันเสาร์จะมีลูกค้ามากที่สุด แต่ค่าอาหารเฉลี่ยของวันอาทิตย์มากกว่า จึงสรุปว่าวันอาทิตย์ทำกำไรเฉลี่ยได้มากกว่าแม้ลูกค้าจะมาน้อยกว่าวันเสาร์

#barplot
#ดูยอดรวมค่าอาหารวันไหนมากที่สุด จากกราฟก่อนหน้าจะเป็นเพียงยอดอาหารเฉลี่ย หากคิดยอดรวมของวัน
pvt = df.pivot_table(index = 'day' , values= 'total_bill', aggfunc=['mean','min','max','sum','std'])
pvt

#ใช้ groupby ได้เช่น
#df.groupby('day').total_bill.mean().sort_values(ascending =False)
#df.groupby('day').total_bill.min().sort_values(ascending=False)
df.groupby('day').total_bill.sum().sort_values(ascending =False)
df.groupby('day').total_bill.sum().sort_values(ascending=False).reset_index()
#การ group วัน แล้วเก็บค่าตัวแปรเพื่อนำไปใช้
group_data = df.groupby('day').sum().reset_index() #กรุ๊ปตามวัน
group_data = df.groupby(['day','sex']).sum().reset_index() #กรุ๊ปตามวัน แล้ว sex
group_data = df.groupby('day').sum().sort_values('total_bill').reset_index() #เรียงข้อมูล
group_data #ดูข้อมูล

group_data.plot.bar() # พล็อตกราฟ
#สรุป การรวบยอด สามารถใช้ได้ทั้ง Pivot Table และ groupby โดยถ้าคำนวณหลายๆค่า เช่น sum mean max min std ใช้ Pivot จะสะดวกกว่า

#กราฟดูยอดรวม
pvt = df.pivot_table(index='day', values='total_bill', aggfunc=['sum'])
#pvt.columns = pvt.columns.droplevel() # drop 1 level
pvt.columns = pvt.columns.get_level_values(1)

pvt.plot.bar()
plt.title('ค่าอาหารรม Bill-Sex')

#ถ้าใช้ Seaborn
sns.barplot(pvt.index,'total_bill',data=pvt)

group_data = df.groupby('day').sum().reset_index() # รวม กรุ๊ป โดย day

sns.barplot(x='day', y='total_bill',data=group_data, palette='Set2') # palette='Set2'รูปแบบสี
sns.lineplot(x='day',y='total_bill',data=group_data, marker = 's', color = 'r')

for index,row in group_data.iterrows():
    plt.annotate(round(row.total_bill,2),    #ข้อความกำกับ
                (row.name, row.total_bill),  #จุดข้อมูล
                textcoords="offset points",  #offset ตำแหน่งข้อความ
                xytext = (0,2),              #ระยะจากข้อความ ถึงจุด points
                ha='center',                 #horizontal alignment = center
                fontsize=15)
plt.title('ยอดขายรวม(Total Bill) by Day')
plt.show()

#ชายหญิงใครจ่ายมากกว่ากัน
pvt = df.pivot_table( index = 'sex',values ='total_bill', aggfunc= ['mean','min','max','sum','std'])
pvt

# กราฟ ยอดค่าอาหารรวมและค่าเฉลี่ย
pvt = df.pivot_table(index= 'sex',values = 'total_bill',aggfunc= ['sum'])
pvt.columns = pvt.columns.get_level_values(1)
pvt.plot.bar()
plt.title('ค่าอาหารรวม Bill-Sex')
plt.title('ค่าอาหารเฉลี่ย Bill-Sex')
plt.show()
sns.barplot('sex','total_bill',data = df)
plt.show()

#จำนวนลูกค้าแบ่งตามเพศ

df.groupby('sex').total_bill.count()
sns.countplot('sex',data=df)
plt.title('จำนวนลูกค้าแบ่งตาม Sex')
plt.show()

#จัดโต๊ะแบ่งโซน
# smk = df.groupby(['smoker']).total_bill.sum().reset_index()  
# ใช้ pivot_table
pvt = df.pivot_table(index='smoker',values = 'total_bill', aggfunc='sum').reset_index()
pvt
pvt.plot.bar()
plt.show()
#sns.barplot('smoker','total_bill',data=pvt) #seaborn
#จะได้ว่าโซนไม่สูบบุหรี่มากกว่าโซนสูบบุหรี่

#ลองคิดยอดรวมแยกพิจารณาเป็นรายวัน และแบ่งตามโซนบุหรี่
# คำนวณด้วย pivot_table
pvt = df.pivot_table(index = 'day', columns ='smoker', values = 'total_bill', aggfunc = 'count')
pvt
# สำหรับกราฟ
sns.countplot(df.day, hue=df.smoker)
plt.legend(loc = 'lower right') #ตำแหน่ง legend กำหนด loc = best ให้เครื่องเลือกวางให้
plt.title('สรุป zone-day')
plt.show()
#สรุปจากกราฟได้ว่า โซนไม่สูบบุหรี่ทำรายได้มากกว่า โดยเฉพาะวันพฤหัสบดีและวันอาทิตย์  ส่วนวนศุกร์โซนสูบบุหรี่ทำรายได้มากกว่า

#การพล็อตข้อมูลชนิดต่อเนื่อง
#ถ้าใช้ countplot กับยอดค่าใช้จ่ายกราฟจะถี่มากไม่เหมาะสม
sns.countplot(df.total_bill) # ไม่มีความหมายใดๆสำหรับ datavisualization
plt.show()
#ใช้ distplot แก้ปัญหาด้านบน
sns.distplot(df.total_bill, bins=20 , color='g')
plt.show()

# สรุปกราฟได้ว่า ยอดส่วนใหญ่อยู่ที่ 10-20 USD
#ใช้ DataFrame.hist
df.total_bill.hist(bins=20)
df.hist('total_bill', bins=20) #ได้กราฟคล้ายกันแค่ต่างที่style-seaborn Vs Pandas(matplotlib)

#kdeplot(Kernel density estimation) คล้าย distplot
sns.kdeplot(df.total_bill, shade=True)
```

```python
#การเปรียบเทียบโดยใช้ Facetgrid
#ถ้าต้องการจะดูลึกไปอีกว่าแต่ละมื้อนั้นแต่ละโซน (สูบบุหรี่-ไม่สูบ)เป็นอย่างไรบ้าง โซนใดทำรายได้มากกว่า โดยใช้ Facetgrid ซึ่งเป็นการแบ่งกราฟออกเป็นแถวและคอลัมน์
import seaborn as sns
sns.set_style('darkgrid') # ใช้รูปแบบกราฟ พื้นสีเทา
grid = sns.FacetGrid(df, row='smoker', col='time')     # row=แถว col=คอลันม์
grid.map(plt.hist,'total_bill', bins=10)
plt.show()
#grid.map(sns.distplot, 'total_bill', kde=False) #ใช้ seabon ก็ได้
#จากกราฟ มื้อกลางวัน ทั้งสองโซนมีค่าเฉลี่ยยอดค่าอาหารพออกัน ส่วนใื้อเย็นโนบุหรี่ ดูเหมือนว่าจะมีค่าเฉลี่ยทำรายได้มากกว่า ซึ่งกราฟนี้อาจเปรียบเทียบยาก ให้ดู Boxplot ถัดไป
```

```python
#การเปรียบเทียบโดยใช้ Boxplot
#เราสามารถใช้ boxplot เปรียบเทียบข้อมูลที่สนใจได้ ตัวอย่างเช่น เปรียบเทียบยอดจ่ายค่าอาหารเฉลี่ยต่อใบเสร็จ ระหว่าง มื้อกลางวัน กับมื้อเย็น โดยกรณีนี้ให้ time เป็นแกน x (เน้นว่าแกนx ควรเป็นข้อมูลชนิดไม่ต่อเนื่อง)
#หรือเป็นแบบนามบัญญัติ หรือเรียงอันดับไม่ใช่ทศนิยม ส่วนแกน y ในกรณีนี้เป็นยอดชำระ (เป็นข้อมูลชนิดต่อเนื่อง)
#sns.boxplot(df.time, df.total_bill, width=0.4) # width=ความกว้างแท่งกราฟ
#sns.boxplot(x='time',y='bill_total', data=df, width=0.4, palette='husl')
#สามารถจัดเรียงลำดับแท่ง box ได้โดยใส่ order
#sns.boxplot(df.time, df.total_bill, order=['Dimmer','Lunch'])
#แบ่งได้ hue ได้เช่น
#sns.boxplot(df.time, df.total_bill, hue=df.smoker, width=0.4, palette='husl') #,order = ['Dinner','Lunch']
#ได้  Boxplot มื้อเย็นโซนสูบบุหรี่ข้อมูลมีการกระจายขึ้นไปสูงกว่า
#ข้อมูลสรุป Q1 Q2 Q3 สำหรับ Boxplot

#ต้องการค่าตัวเลขสรุป Q1 Q2 Q3 สำหรับ ฺBoxplot ให้ใช้วิธีการรวมสรุป groupby (quantile .25 .5 .75) คือ Q1 Q2 Q3 ของboxplot
df.groupby(['time']).total_bill.quantile([.25,.5,.75]).to_frame() #ได้สรุปแบบแนวดิ่ง
df.groupby(['time']).total_bill.quantile([.25,.5,.75]).to_frame().T #ได้้สรุปแนวนอน
#จากที่ผ่านมาเป็นการแยกมื้ออาหาร ในกรณีต้องการเปรียบเทียบระหว่างโซนสูบบุหรี่และไม่สูบบุหรี่ โดยไม่สนใจมื้ออาหาร ใช้โค้ดด้านล่าง จะได้ว่าค่าเฉลี่ยของโซนบุหรี่ต่อใบเสร็จบิล สูงกว่าโซนไม่สูบบุหรี่
#sns.boxplot(df.smoker,df.total_bill, width=0.5)

#ต้องการเปรียบเทียบรายวัน โซนสูบบุหรี่กับไม่สูบบุหรี่
sns.boxplot(df['day'], df.total_bill,width =0.6, hue=df.smoker)
plt.legend(loc='upper left')
#พบว่าเสาร์อาทิตย์ โซนสูบบุหรี่จ่ายค่าอาหารเฉลี่ยสูงกว่า ค่าเฉลี่ยเท่านั้น

#คิดยอดรวม
#หัวข้อนี้จะพิจารณายอดรวมรายรับทั้งหมด(รวมยอดทุกใบเสร็จ)
pvt = df.pivot_table(index='day' ,columns='smoker',values='total_bill',aggfunc=['sum','count'])
pvt.plot.bar()
pvt
#จากBoxplot จะเห็นว่าโซนสูบบุหรี่วันอาทิตย์จะมีปริมาณมากกว่าแต่นั่นคือค่าเฉลี่ยใบเสร็จ (ยอดชำระต่อใบเสร็จ) แต่พอพิจารณากราฟยอดรวมของโซนไม่สูบบุหรี่ สูงกว่าอย่างเห็นได้ชัด
#เพราะจำนวนลูกค้าโซนไม่สูบบุหรี่มีสูงกว่า
#สรุปได้ว่า กลุ่มลูกค่าสูบบุหรี่วันอาทิตย์มีน้อยกว่า แต่จ่ายบิลเฉลี่่ยหนักกว่า ส่วนกลุ่มไม่สูบบหรี่ มียอดชำระรวมสูงกว่า คือจ่ายโต๊ะเฉลี่ยต่ำกว่าของจริง แต่จำนวนคนเยอะกว่ายอดรวมเลยมากกว่า
#ดังนั้นการทำdatavisualization จึงควรดูให้ครบถ้วนในทุกมิติ จึงจะได้ข้อสรุปที่แม่นยำและใช้ประโยชน์ได้ในธุรกิจ
```

```python
#กราฟ Violinplot
#กรามีลักษณะคล้ายไวโอลิน โดยจะแสดงการกระจายตัวเหมือนกับำ Boxplot + Displot หรือ KDE plot แสดงค่าการกระจาย 2 ข้าง
sns.violinplot(df.total_bill) #แบบ1
plt.show()
sns.violinplot(df['total_bill']) #แบบ2
plt.show()
sns.violinplot('total_bill', data = df) # แบบ3
plt.show()
#รูปโค้งกราฟแสดงค่าการกระจายคลา้ยๆ ฮิสโทแกรม และKDE plotซึ่งตีความหมายรวมๆคือ ค่าเฉลี่ยยอดชำระอยู่ที่ประมาณ 18-19 ต่อใบเสร็จ เมื่อเทียบ violinplot กับ displot จะได้รูปทรงประมาณเดียวกัน ค่าที่เท่าๆกัน(แตกต่างที่ความสูงของกราฟเท่านั้น แต่ค่าต่างๆเท่ากัน)
#หากต้องการกราฟแนวตั้ง ให้กำหนด y =ข้อมูล
sns.violinplot(y=df.total_bill) # แบบ 1
plt.show() 
sns.violinplot(y='total_bill',data=df) # แบบ 2
plt.show()
#ต้องการเปรียบเทียบระหว่าง ยอดรวม กับ เวลามื้ออาหาร
sns.violinplot(x= 'time', y='total_bill', data=df) #ได้แบบแนวตั้ง
plt.show()
sns.violinplot(y= 'time', x='total_bill', data=df) #ได้แบบแนวนอน
plt.show()
#ต้องการดูยอดจ่ายตามวัน
sns.violinplot(x='day',y='total_bill',data =df) #แบบไม่จัดเรียง
plt.show()
ord = ['Fri','Thur','Sat','Sun'] #แบบจัดเรียงตามวัน ใช้ order =  ตัวแปรที่จะจัดเรียง
sns.violinplot('day','total_bill',data= df,order = ord)
plt.show()

#ต้องการพล็อตแยกกลุ่มโซนสูบบุหรี่และไม่สูบบุหรี่แบ่งตามวัน
sns.violinplot('day','total_bill',data=df, hue = 'smoker') #ใช้ ้ hue= เข้าช่วย
plt.show()
#Violinplot แบ่ง 2 ซีก
#จุดเด่นของ violinplotนั่นคือสามารถแบ่งได้ 2 ซีก จะยุบกราฟรวมแล้วแบ่งเป็น 2 ซีก
sns.violinplot('day','total_bill',data=df , order = ord , hue='smoker', split=True) #ใช้ split=True
#จะได้กราฟซีกซ้ายไม่สูบบุหรี่ ขวาสูบ และจะยุบรูปจาก 8 แท่งเหลือ 4 แท่ง
```
