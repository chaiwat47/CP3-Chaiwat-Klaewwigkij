```python
import pandas as pd
import numpy as np
```

```python
#สร้าง series จาก Numpy
datas = [15,12.5,'Ksb',22,"Jim"] #สร้าง List ก่อน
ndata = np.array(datas) #นำList ไปสร้างเป็น Numpy Array
ps = pd.Series(ndata) #สร้าง Series จาก Numpy
ps
```

```python
#สร้างข้อมูล series แบบกำหนด index
datas = [15,12.5,18,22,17] #สร้างข้อมูล list
idx = ['tide','ไทด์','time','แมว','dog'] #index
ps = pd.Series(datas, index=idx) #สร้างSeriesแบบกำหนด index เอง
ps
```

```python
#สร้างข้อมูล series แบบกำหนด index โดยใช้numpy
datas = [15,12.5,18,22,17] #สร้างข้อมูล list
idx = ['tide','ไทด์','time','แมว','dog'] #
ps = pd.Series(np.array(datas), index=idx) #สร้างSeriesแบบกำหนด index โดยแปลงdatas เป็น array โดยnumpy
ps
```

```python
#สร้าง series จาก Dictionary
datas = {'tide':15,'ไทด์':12.5,'time':18,'แมว':22,'dog':17} #รวบข้อมูล:index โดยใช้Dictionary
ps = pd.Series(datas)
ps
```

```python
#การเข้าถึงข้อมูล สามารถทำได้ 2 วิธี คือ 1.อ้างอิงลำดับ index หรือ 2.อ้างอิงชื่อ index
ps[0] #อ้างอิงลำดับ
ps[1] #อ้างอิงลำดับ
ps[-1] #อ้างอิงลำดับ
ps['tide'] #อ้างอิงชื่อ
ps['ไทด์'] #อ้างอิงชื่อ
```

```python
#การเข้าถึงช่วงของข้อมูล (Slicing) ทำโดยใช้ series[ indexเริ่ม : indexสุดท้าย ]
ps[:] #เอาทั้งหมด
ps[2:] #ตั้งแต่index2เป็นต้้นไป
ps[:2] #ตั้งแต่0-2(ที่index2ไม่เอา)
ps[1:3] #ตั้งแต่1-3 (ที่ 1 เอา 3ไม่เอา)
ps[2:len(ps)] #ตั้งแต่2-ความยาวของseriesจบ
```

```python
#การสร้าง DataFrame
#-สร้างจาก Tuple or List
datas = [15,12.5,18,22,17] #ข้อมูลที่ใช้ต้องเป็นlist
cols = ['Age'] #สร้างlist เก็บชื่อcolumn
df = pd.DataFrame(datas, columns=cols) #สร้างdataframe พร้อมระบุชื่อcolumn
df
# -สร้างจาก Tuple or List จากหลายมิติ หรือ เมทริกซ์
datas = [[163,37,1],
         [175,20,0],
         [162,27,0],
         [150,10,1]]
cols = ['height','hair_length','voice_pitch']
df = pd.DataFrame(datas, columns=cols)
df
#สร้าง DataFrame จาก List หลายตัวโดยใช้Zip
#บางข้อมูลมีlistย่อยหลายตัวต้องทำการzipก่อน จึงทำขึ้นเป็นdataframeได้
lst1 = [15,22,30,24]
lst2 = [20,18,38,19]
datas = list(zip(lst1,lst2)) #ใช้คำสั่ง zip()เพื่อจับคู่ข้อมูล[(15, 20), (22, 18), (30, 38), (24, 19)]

cols = ['Age','Score']
df = pd.DataFrame(datas, columns = cols)
df

#สร้าง DataFrame แบบ กำหนดชื่อ Columns
#lst0 = ['tide','jim','god']
#lst1 = [15,22,24]
#lst2 = [20,18,19]
#datas = list(zip(lst0,lst1,lst2))
datas = [['tide',15,20],['jim',22,18],['god',24,19]]
cols =['Name','Age','Score']
df = pd.DataFrame(datas, columns = cols)
df

#สร้าง DataFrame จาก Dictionary
datas = [{'Name':'tide','Age':15,'Score':20},
        {'Name':'jim','Age':22,'Score':18},
        {'Name':'god','Age':24,'Score':19}]
df = pd.DataFrame(datas)
df

#สร้าง DataFrame จาก series
names = ['tide','jim','god']
ages = [15,22,24]
scores = [20,18,19]
cols =['Name','Age','Score']
name_series = pd.Series(names)
age_series = pd.Series(ages)
score_series = pd.Series(scores)

frame = {'Name':name_series,'Age':age_series,'Score':score_series} #เก็บseries พร้อมระบุ columns
df = pd.DataFrame(frame)
df
#กรณีที่มี Series อยู่ columnsเดียวต้องการเปลี่ยนเป็นDataFrame ให้ใช้คำสั่ง .to_frame()
#df = name_series.to_frame(name = 'StudentName') #ตั้งชื่อcolumn ใช้ name = ...
#df
```
