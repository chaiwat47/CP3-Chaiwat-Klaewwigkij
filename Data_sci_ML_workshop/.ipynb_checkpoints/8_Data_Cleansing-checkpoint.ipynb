# Workshop 7 : คัดตัวเลขออกจาก Object / String 

```python
df = pd.read_csv('data_clean2_after_str_upper.csv')
print(df.dtypes)
df

#ฟังก์ชันตรวจจับตัวเลข ให้เปลี่ยนเป็น NaN (เราต้องการ Str ไม่ต้องการตัวเลข)
def detect_num(dataframe, column_name):
 cnt = 0
 for row in dataframe[column_name]:
     try:
         # int(row)
         float (row)
         dataframe.loc[cnt, column_name] = np.nan
     except ValueError:
         pass
     cnt+=1
 return(dataframe)
#
df = detect_num(df.copy(),'own')
df
df.dropna(subset=['own'], inplace=True) #ลบแถวที่มี NaNออก
df
```

# Workshop 10 : จัดการข้อมูลตัวเลข

```python
# จะเห็นว่า bedroom and bathroom
# bedroom จำนวนห้องนอนไม่ควรเป็นเลขทศนิยม ควรเป็นจำนวนเต็ม
# bathroom จำนวนห้องน้ำไม่ควรติดลบ ทศนิยม ไม่มีข้อความ
df.bathroom = pd.to_numeric(df.bathroom, errors='coerce', downcast = 'integer')
df.dtypes
df
# รูปแบบ 1 (เขียนเต็ม)
def float2int(val):
    if val % 1 == 0 : #จำนวนที่อยู่ใน bathroom มาหาร 1 แล้วต้องเศษเหลือ 0 (หารไม่เหลือเศษ)
        return int(val) #แปลงเป็นจำนวนเต็ม
    else:
        return np.nan #ถ้าไม่ใช่ตามเงื่อนไขให้ใส่NaN
df.bathroom = df.bathroom.apply(float2int) #เรี่ยกใช้ฟังก์ชันโดยเอาค่าจาก bathroom ไปใส่ลงในฟังก์ชันตรวจสอบเงื่อนไข
# df['bathroom_num'] =df['bathroom_num'].apply(float2int) #แบบนี้ก็ได้ สร้างbathroom_numรอรับค่า

# รูปแบบ2
#def float2int(val):
#    return int(val) if val%1 == o else np.nan
#df.bathroom = df.bathroom.apply(float2int) # เรียกฟังก์ชันเปลี่ยน

# รูปแบบ 3 การใช้ lambda
#df.bathroom = df.bathroom.apply(lambda x: x if (x%1==0)else np.nan)
df
#ถ้าต้องการล้าง NaN 
df = df[df.bathroom.notnull()]
df
df.bathroom = df.bathroom.astype(int)
df.dtypes
df
#ยังเหลือส่วนที่ติดลบอยู่
(df.bathroom <= 0).sum() #ดูว่ามีค่าน้อยกว่า 0 กี่ตัว
df.groupby('bathroom').count()
df[df.bathroom <=0]
df = df[df.bathroom>0] #คัดกรองเอาแต่ที่มีค่ามากกว่า 0
df
#ทำเช่นเดียวกับ bathroom ใน bedroom category
# ไม่ให้ bathroom คำนวนสถิติ  
df.bathroom = df.bathroom.astype('category')
```
