# Workshop 1 : โหลดข้อมูลและตรวจสอบเบื้องต้น

```python
import pandas as pd
import numpy as np
#โหลดไฟล์จากCSV

df = pd.read_csv('data_missing.csv',encoding='utf-8')
df.shape #ตรวจสอบขนาด (แถว,หลัก)
df.dtypes #ตรวจสอบชนิดข้อมูล
df.head()
df.tail()
df.sample(5)
df.describe()

%matplotlib inline
import matplotlib.pyplot as plt
df.plot.bar()
plt.show()
df.sample(5)
df.head(10)
```

# Workshop 2 : เปลี่ยนชนิดข้อมูลให้เหมาะสม

```python
#จากผลการคำนวณค่าสถิติเบื้องต้น df.describe() พบว่าวันที่เป็นข้อมูล int64 ซึ่งไม่ควรนำมาหาค่าเฉลี่ยจะต้องเปลี่ยนเป็นชนิดที่ถูกต้อง

df.Date = df.Date.astype('category')
df.describe() # Dateไม่ถูกนำมาคำนวณแล้ว
df.dtypes
#ถ้าเปลี่ยนไม่ได้ก็ไม่ส่งผลต่อ missing data เนื่องจากมีข้อมูลครบ
```

# Workshop 3 :หา Missing data รวมๆทั้งหมด

```python
df.isnull() #ตรวจสอบ missing data โดยรวม
df.isnull().any() # ตรวจดูว่าแถวใดมี missing data บ้าง ถ้ามีตอบ True ไม่มีจะตอบ False
df.isnull().sum() # นับรวมจำนวนข้อมูลให้ ที่มี missing data
df.notnull().sum() # ตรวจสอบข้อมูลที่ไม่ใช่ missing data มีเท่าไหร่

# ตรวจสอบเปอร์เซ็นต์ของ Missing data
missing_count = df.isnull().sum() # จำนวนmissing data

total_cells = np.product(df.shape) # ผลคูณ ช่องข้อมูล = column * row
total_missing = missing_count.sum() # รวมจำนวนทั้งหมด
missing_percent = total_missing/total_cells *100

print('Total cells : {}'.format(total_cells))
print('Total missing values : {}'.format(total_missing))
print('Missing data : {}%'.format(missing_percent))
```

# Workshop 13 : จัดการลบคอลัมน์ ที่มี missing data

```python
df = pd.read_csv('data_missing.csv', encoding = 'utf=8')
df
df.dropna(axis='columns',inplace=True)
df

#จะเห็นได้ว่า column Temp จะถูกลบทิ้งเนื่องจากมีข้อมูล ที่ไม่เป็น missing data เหลือน้อยกว่า 8
```

# Workshop 14 : การลบแบบมีเงื่อนไข

```python
df = pd.read_csv('data_missing.csv',encoding = 'utf-8')
df

df.dropna(thresh = 8, axis=1 ,inplace =True) #threshold ต้องการลบคอลัมน์ทิ้ง ถ้าข้อมูลเหลือน้อยกว่า 8
df

#หรืออยากกำหนดแบบแถวใดมี missing data เกิน 20 percent ให้ลบทิ้งไปเลย
df.shape
df.shape[0] #จำนวน row df.shape[1] จำนวน column
left = int(df.shape[0]*0.8)
left #80%ของจำนวนแถว(จำนวนข้อมูลทั้งหมด) จะมีค่าเป็น 8
df.dropna(thresh = left,axis=1, inplace =True) #เท่ากับลบ thresh = left ถ้ามีข้อมูลเพิ่มให้ลบออกเลย
df
```

# Workshop 15 : การลบถ้าทั้งแถวเป็น Missing data

```python
df.loc[[0,4],:] = np.nan  #ทำให้แถว 0 และ 4 เป็น NaN โดยกำหนดเข้าไปตรงๆ
df
df[df.isnull().any(axis=1)] #หาข้อมูลที่เป็น missing data บางคอลัมน์
df[df.isnull().all(axis=1)] #หาข้อมูลที่เป็น missing data ทั้งคอลัมน์
df.dropna(how = 'all' ,inplace =True)
df
```

# Workshop 16 : ลบ ถ้าทั้งคอลัมน์เป็น missing data

```python
df['location'] = np.nan #เพิ่ม column location ลงไปเป็น NaN
df
df.dropna(axis=1 , how='all', inplace = True) #ใช้ axis =1 คือ ตรวจcolumn ถ้า ไม่ใช้คือ ตรวจ row
df
```
